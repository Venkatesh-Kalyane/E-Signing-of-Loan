```python
df=pd.get_dummies(df)
df
```

```python
df = df.drop(columns = ['months_employed','pay_schedule_monthly','entry_id'])
df['personal_account_months'] = (df.personal_account_m + (df.personal_account_y * 12))
df = df.drop(columns = ['personal_account_m', 'personal_account_y'])
df
```

```python
X=df.drop(['e_signed'],axis=1)
X
```

```python
Y=df['e_signed']
Y
```

```python
from sklearn.model_selection import StratifiedKFold
skf=StratifiedKFold(n_splits=2)
skf.get_n_splits(X, Y)
```

```python
for train_index, test_index in skf.split(X, Y):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = Y.iloc[train_index], Y.iloc[test_index]
```

```python
from sklearn.ensemble import RandomForestClassifier
clf=RandomForestClassifier()
model=clf.fit(X_train,y_train)
model
```

```python
y_pred=model.predict(X_test)
y_pred
```

```python
model.score(X_test,y_test)
```

```python
from sklearn.model_selection import GridSearchCV
params = {
    "n_estimators" : [10, 100, 200,300 ],  "criterion":["entropy", "gini"]
}
gs = GridSearchCV(estimator=clf, param_grid=params, cv=10)
```

```python
result = gs.fit(X_train, y_train)
```
